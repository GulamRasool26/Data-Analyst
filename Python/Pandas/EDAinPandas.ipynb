# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('world_population (1).csv')
df.head()
```

***formatting decimal places in Float numbers***

```python
pd.set_option('display.float_format',lambda x: '%.2f' % x)
```

```python
df
```

***Getting info of DataFrame like data type of columns or null values***

```python
df.info()
```

***Getting Basics Statistical Information of the DataFrame***

```python
df.describe()
```

***Total Number of Null Values in Each rows***

```python
df.isnull().sum()
```

***Finding Unqiueness and Duplicates***

```python
df.nunique()
```

***Sorting Values by 2022 Population***

```python
df.sort_values(by="2022 Population").head(10)
```

***Rearranging top 10 by Descending***

```python
df.sort_values(by="2022 Population", ascending=False).head(10)
```

***top 10 by World Percentage***

```python
df.sort_values(by="World Population Percentage", ascending=False).head(10)
```

***Finding Correlation between Each Column***

```python
df.corr(numeric_only=True)
```

## ***Visualization of Correlation***

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.rcParams['figure.figsize'] = (15, 6)
plt.show()
```

***Analyzing Higher level like Continent Level.***

```python
df.groupby('Continent').mean(numeric_only=True)
```

```python
df.columns
```

```python
df2 = df.groupby('Continent')[['1970 Population',
    '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
```

```python
df[df['Continent'].str.contains('Oceania')].count()
```

```python
df2
```

```python
df3 = df2.transpose()
```

```python
df3.plot()
```
