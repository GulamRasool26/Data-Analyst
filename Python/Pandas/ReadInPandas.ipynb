# **Reading Files Using Pandas**

```python
import pandas as pd
```

### *Reading Different Formats*

```python
pd.read_csv('countries of the world.csv') #csv file
```

```python
pd.read_csv('countries of the world.csv', header=None) 
#csv file without header row, so pandas will assign default column names (0, 1, 2, etc.)
```

```python
pd.read_csv('countries of the world.csv', header=None, names=['Country' , 'Region'])
#csv file without header row, so pandas will assign custom column names (Country, Region, etc.)
```

```python
df = pd.read_csv('countries of the world.csv')
# read the CSV file into a DataFrame and assign it to the variable
df
```

```python
df = pd.read_csv('countries of the world.txt')
# read the tab-delimited text file into a DataFrame and assign it to the variable
df
```

```python
df = pd.read_csv('countries of the world.txt', sep='\t')
# read the tab-delimited text file into a DataFrame, specifying the separator as a tab character
df
```

```python
df = pd.read_table('countries of the world.txt')
# read the tab-delimited text file into a DataFrame using read_table, which defaults to tab as the separator
df
```

```python
df = pd.read_table('countries of the world.csv', sep=',')
# read the CSV file into a DataFrame using read_table, specifying the separator as a comma character
df
```

```python
df = pd.read_json('json_sample.json')
# read the JSON file into a DataFrame and assign it to the variable
df
```

```python
df = pd.read_excel('world_population_excel_workbook.xlsx')
# read the Excel file into a DataFrame and assign it to the variable
df
```

```python
df = pd.read_excel('world_population_excel_workbook.xlsx' , sheet_name='Sheet1')
# read the Excel file and specify the sheet name to read from
df
```

### *Setting Outputs rows and columns Limit*

```python
pd.set_option('display.max.rows',235)
# set the maximum number of rows to display in the output to 235
```

```python
pd.set_option('display.max.columns',40)
# set the maximum number of columns to display in the output to 40
```
