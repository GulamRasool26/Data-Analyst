## Comparison 

| Operator | Meaning                     |
|----------|-----------------------------|
| ==       | Equal                       |
| !=       | Not Equal                   |
| >        | Greater than                |
| <        | Less than                   |
| >=       | Greater than or equal to    |
| <=       | Less than or equal to       |

```python
10 == 50
```

```python
10 != 50
```

```python
'Vanilla' == 'Vanilla'
```

```python
x = 'Vanilla'
y = 'Chocolate'
x == y
```

```python
x != y
```

```python
10<=10
```

```python
10>=10
```

```python
50>=10
```

## Logical Operators

| Operators | Description                        |
|-----------|------------------------------------|
| and       | Returns True if Both are True      |
| or        | Returns True if atleast one is True|
| not       | Reverse the result, True if False  | 

```python
(10 > 50) and (50 > 10 )
```

```python
(70 > 50) and (50 > 10 )
```

```python
(10 > 50) or (50 > 10 )

```

```python
('Vanilla' > 'Chocolate') and (10 < 50)
```

```python
not (10 > 50) and (50 > 10 )
```

## Membership Operators

| Operators | Description                                                                      |
|-----------|----------------------------------------------------------------------------------|
| in        | Returns True if a sequence with specified value is PRESENT in the Object         |
| not in    | Returns True if a sequence with the specified value is not [resent in the object |


```python
ice_cream = "I love ice cream"
'love' in ice_cream
```

```python
scoops =  [1,2,3,4,5]
3 in scoops
```

```python
6 not in scoops
```
