# **Group by and Aggregation**

```python
# importing pandas library 
import pandas as pd
```

```python
# reading the csv file and creating a dataframe
df = pd.read_csv('Flavors.csv')
df
```

```python
# getting the information about the dataframe
df.info()
```

```python
# grouping the dataframe by 'Base Flavor' column
grouped_df = df.groupby('Base Flavor')
```

```python
# calculating the mean of the grouped dataframe
grouped_df.mean(numeric_only=True)
```

```python
# performing aggregation on the grouped dataframe
df.groupby('Base Flavor').mean(numeric_only=True)
```

```python
# getting the count of each group
grouped_df.count()
```

```python
# getting the minimum value of each group
grouped_df.min()
```

```python
# getting the maximum value of each group
grouped_df.max()
```

```python
# getting the sum of each group
grouped_df.sum()
```

```python
# performing multiple aggregation on the grouped dataframe
grouped_df.agg({'Flavor Rating':['mean','max','count','sum']})
```

```python
# grouping the dataframe by 'Base Flavor' and 'Liked' columns
df.groupby(['Base Flavor','Liked']).mean(numeric_only=True)
```

```python
# getting the descriptive statistics of each group
df.groupby('Base Flavor').describe()
```
