# **Indexes**

```python
# importing the pandas library
import pandas as pd
```

```python
# reading the csv file and setting the index to the 'Country' column
df = pd.read_csv('world_population.csv', index_col = 'Country')
df
```

```python
# resetting the index to the default integer index
df.reset_index(inplace = True)
df
```

```python
# setting the index to the 'Country' column again
df.set_index('Country', inplace = True)
```

```python
df
```

### *Searching through Index*

```python
# searching for the population of Albania using the index string
df.loc['Albania']
```

```python
# searching for the population of the second country in the dataframe using the default integer index
df.iloc[1]
```

```python
# resetting the index in place
df.reset_index(inplace = True)
df
```

```python
# setting a multi-index using the 'Continent' and 'Country' columns
df.set_index(['Continent' ,'Country'], inplace = True)
```

```python
df
```

```python
# sorting the multi-index in place
df.sort_index(inplace = True)
df
```

```python
# searching for the population of Angola in Africa using the multi-index
df.loc['Africa','Angola']
```

```python
# searching for the population of Albania using the default integer index
df.iloc[1]
```
