# **Web Scraping 3**
### *Find and Find_all*

```python
# importing the libraries
import requests
from bs4 import BeautifulSoup
```

```python
# defining url
url = "https://www.scrapethissite.com/pages/forms/"
```

```python
# making a get request
response = requests.get(url)
```

```python
# parsing the response
soup = BeautifulSoup(response.text, "html.parser")
```

# **Querying the HTML**


```python
soup.find('div') # it will return the first div element
```

```python

soup.find_all('div') # it will return all the div elements in the form of list
```

```python
soup.find_all('div', class_='col-md-12') # it will return all the div elements with class name col-md-12 in the form of list
```

```python
soup.find_all('p') # it will return all the p elements in the form of list
```

```python
soup.find_all('p',class_='lead') # it will return all the p elements with class name lead in the form of list
```

```python
soup.find_all('p',class_='lead').text # it will not return the text of all the p elements with class name lead in the form of list
                                    # because we cannot use text attribute with find_all method. We have to use it with find method only. 
                                    # So we can use a loop to get the text of all the p elements with class name lead in the form of list
```

```python
soup.find('p',class_='lead').text # it will return the text of the first p element with class name lead
```

```python
soup.find('p',class_='lead').text.strip() # it will return the text without any leading and trailing whitespace characters
```

```python
soup.find_all('th') # it will return all the th elements in the form of list
```

```python
soup.find_all('td') # it will return all the td elements in the form of list
```

```python
soup.find('th').text.strip() # it will return the text of the first th element without any leading and trailing whitespace characters
```
