# Data Cleaning in Pandas

```python
import pandas as pd 
```

Reading Files

```python
df = pd.read_excel('Customer Call List.xlsx')
df.head()
```

***Finding Number of Null Values***

```python
df.info()
```

***droping Duplicates rows***

```python
df = df.drop_duplicates()
```

***droping Unnecessary Columns***

```python
df = df.drop(columns= 'Not_Useful_Column')
df
```

***Formatting Columns ***

```python
df['Last_Name'] = df['Last_Name'].str.lstrip("...")
df['Last_Name'] = df['Last_Name'].str.lstrip("/")
df['Last_Name'] = df['Last_Name'].str.rstrip("_")
df['Last_Name']

```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace("-", "")
df['Phone_Number'] = df['Phone_Number'].str.replace("/", "")
df['Phone_Number'] = df['Phone_Number'].str.replace("|", "")
df.head()
```

```python
#convert the datatype of phone number to string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x : x[0:3] + "-" + x[3:6]+"-"+ x[6:10])
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace("nan--", "")
df['Phone_Number'] = df['Phone_Number'].str.replace("Na--", "")
```

```python
df.head()
```

***Splitting Column using delimiters***

```python
df['Address'].str.split(",", expand = True)
```

***Creating New Features By splitting a single Column to three New Columns***

```python
df[['Street','State', 'Zip_Code']] = df['Address'].str.split(",", expand = True)
df.head()
```

***Combine Cleaning Because Same type of Binary labels***

```python
df[['Paying Customer','Do_Not_Contact']]
```

```python
df[['Paying Customer','Do_Not_Contact']] = df[['Paying Customer','Do_Not_Contact']].replace({'Yes': 'Y', 'No':'N'})
```

```python
df.head()
```

```python
df = df.replace('N/a','')
df = df.replace('NaN','')
```

```python
df
```

***Filling Null Values***

```python
df = df.fillna('')
```

***looping the Dataframe to remove Unnecesary rows on basis of Condition***

```python
for x in df.index:
    if df.loc[x,'Do_Not_Contact'] == 'Y':
        df.drop(x,inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x,'Phone_Number'] == '':
        df.drop(x,inplace=True)
df
```

***Reseting Index after deleting rows***

```python
df =df.reset_index(drop=True)
```
