# Filtering And Ordering in Pandas

```python
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv('world_population.csv')
df
```

### *Filtering*

```python
# Filter the dataset to include only the top 10 countries by population
df[df['Rank'] <= 10] #numeric filter
```

```python
# Filter the dataset to include only countries in List
specific_countries = ['Bangladesh','Brazil']
df[df['Country'].isin(specific_countries)] #filtering using a list of values
```

```python
# Filter the dataset to include only countries that contain 'United' in their name
df[df['Country'].str.contains('United')] # filtering using string methods
```

```python
# Set 'Country' as the index of the DataFrame
df2 = df.set_index('Country')
df2
```

```python
# filtering columns 
df2.filter(items = ['Continent', 'CCA3'])
```

```python
# filtering columns using axis parameter 
df2.filter(items = ['Continent', 'CCA3'], axis = 1) # 1 for columns, 0 for rows
```

```python
df2.filter(items = ['Continent', 'CCA3'],axis=0)
#axis 0 filters rows, so it will look for index labels 'Continent' and 'CCA3' and return those rows if they exist.
#Since 'Continent' and 'CCA3' are not index labels in df2, 
#the result will be an empty DataFrame.
```

```python
df2.filter(items = ['Zimbabwe'], axis=0)#This will return the row corresponding to 'Zimbabwe' 
#if it exists in the index of df2, otherwise it will return an empty DataFrame.
```

```python
df2.filter(like= 'United', axis=0)#This will return all rows where the index label contains the substring 'United'.
```

```python
df2.loc['United States']#This will return the row corresponding to 'United States
#if it exists in the index of df2, otherwise it will raise a KeyError.
```

```python
df2.iloc[3]#This will return the row at the 4th position (index 3) in the DataFrame df2, regardless of the index labels.
```

### *Ordering*

```python
#first filter the DataFrame to include only countries with a 'Rank' less than 10
#then sort the resulting DataFrame by 'Rank' in descending order and by 'Country'
df[df['Rank']<10].sort_values(by=['Rank','Country'], ascending=False)
```

```python
#first filter the DataFrame to include only countries with a 'Rank' less than 10
#then sort the resulting DataFrame by 'Rank' in descending order and by 'Country'
df[df['Rank']<10].sort_values(by=['Rank','Country'], ascending=[False,True])
```
