# **Conditional Statments**
* if
* if - Else
* if - Elif
* Nested IF

```python
if 25 < 10 or 10 > 5:
    print("At least one condition is true")
```

```python
if (25 > 10) or (10 > 20):
    print("At least one condition is true")
```

```python
if 25 < 10:
    print("25 is less than 10")
else:
    print("25 is not less than 10")
```

```python
if 5 < 10:
    print("5 is less than 10")
else:
    print("5 is not less than 10")
```

```python
if 5 > 10:
    print("5 is greater than 10")
elif 5 == 10:
    print("5 is equal to 10")
elif 5!= 10:
    print("5 is not equal to 10")
else:
    print("5 is not greater than 10")
```

```python
print ("5 is less than 10") if 5 < 10 else print("5 is not less than 10")
```

```python
if 5 > 10 or 10 == 10:
    print("5 is greater than 10")
    if 10 > 5:
        print("10 is greater than 5")

elif 5 == 10:
    print("5 is equal to 10")
elif 5!= 10:
    print("5 is not equal to 10")
else:
    print("5 is not greater than 10")
```
