```python
integers = [1, 2, 3, 4, 5]
```

```python
for n in integers:
    print(n)
```

```python
for n in integers:
    print("Hello, World!")
```

```python
for n in integers:
    print( n * n)
```

```python
dict_1 = {'name': 'Gulam' , 'age': 22, 'city': 'Faisalabad', 'skills': ['Python', 'Data Analysis', 'Machine Learning']}
```

```python
for data in dict_1.values():
    print(data)
```

```python
for key, data in dict_1.items():
    print(f"{key}: {data}")
```

## Nested For loops

```python
skills = ['Python', 'Data Analysis', 'Machine Learning']
concepts = ['Data Structures', 'Design Patterns', 'Object-Oriented Programming']
```

```python
for one in skills:
    for two in concepts:
        print(f"{one} - {two}")
    print("-------------")    
```

## WHILE LOOPS
can control using BREAK and Continue 

```python
num = 0
while num < 5:
    print(num)
    num += 1
```

```python
num = 0
while num < 5:
    print(num)
    if num == 3:
        break
    num += 1
```

```python
num = 0
while num < 5:
    print(num)
    if num == 6:
        break
    num += 1
else:
    print("Loop completed without break.")
```

```python
# Infinit loop example (commented out to prevent execution)
num = 0
while num < 5:
    print(num)
    if num == 3:
        continue
    num += 1
else:
    print("Loop completed without break.")
```

```python
num = 0
while num < 5:
    num += 1
    if num == 3:
        continue
    print(num)
else:
    print("Loop completed without break.")
```
