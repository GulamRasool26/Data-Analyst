# Converting Data Types


```python
num_int = 7
type(num_int) # This will return <class 'int'>, which indicates that num_int is an integer.
```

```python
num_str = '7'
type(num_str) # This will return <class 'str'>, which indicates that num_str is a string.
```

```python
num_sum = num_int + num_str # This will raise a TypeError, because you cannot add an integer and a string together. To fix this, you can either convert the integer to a string or the string to an integer before performing the addition.
```

```python
num_str_cnv = int(num_str) # This will convert the string '7' to the integer 7, and store it in the variable num_str_cnv.
type(num_str_cnv) # This will return <class 'int'>, which indicates that
```

```python
num_sum = num_int + num_str_cnv # This will now work, because both num_int and num_str_cnv are integers, so they can be added together without any issues. The result will be 14, which is the sum of 7 and 7.
num_sum
```

## Converting List
* list to tuple
* dict to list
* list to set
* string to list
* string to set

```python
list_type = [1, 2, 3]
type(list_type) # This will return <class 'list'>, which indicates that list_type
```

```python
type(tuple(list_type)) # This will return <class 'tuple'>, which indicates that the tuple() function has successfully converted the list into a tuple. The resulting tuple will contain the same elements as the original list, but it will be immutable, meaning that you cannot change its contents after it has been created.
```

```python
list_type = [1, 2, 3,4,4,5,6,6,7,88,5,90]
```

```python
set(list_type) # This will convert the list into a set, which is an unordered collection of unique elements. The resulting set will contain only the unique values from the original list, so any duplicate values will be removed. In this case, the resulting set will contain the values {1, 2, 3, 4, 5, 6, 7, 88, 90}, since those are the unique values in the original list.
```

```python
type(set(list_type)) # This will return <class 'set'>, which indicates that the set() function has successfully converted the list into a set. The resulting set will contain only the unique values from the original list, and it will be an unordered collection of those unique elements.
```

```python
dict_type = {'name': 'Gulam', 'age': 22, 'hairColor': 'black'}
type(dict_type) # This will return <class 'dict'>, which indicates that dict_type is a dictionary. A dictionary is a collection of key-value pairs, where each key is unique and maps to a specific value. In this case, the dictionary contains three key-value pairs: 'name' maps to 'Gulam', 'age' maps to 22, and 'hairColor' maps to 'black'.
```

```python
dict_type.items()
```

```python
dict_type.values() # This will return a view object that displays a list of the dictionary's key-value pairs as tuples. Each tuple contains a key and its corresponding value from the dictionary. In this case, the output will be dict_values(['Gulam', 22, 'black']), which shows the three values in the dictionary.
```

```python
dict_type.keys()
```

```python
type(list(dict_type.keys())) # This will return <class 'list'>, which indicates that the keys() method of a dictionary returns a view object that displays a list of the dictionary's keys. In this case, the output will be ['name', 'age', 'hairColor'], which shows the three keys in the dictionary.
```
