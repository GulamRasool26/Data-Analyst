## Parametrized Functions


```python
# num is the parameter for the function, and it will be squared when the function is called. 
def num_squared(num):
    print(num ** 2)
```

```python
num_squared(4) # This will print 16, which is 4 squared.
```

```python
# multiple parameters
def num_squared(num, power):
    print(num ** power)
```

```python
num_squared(4,3)
```

### Default Parameters

```python
# default parameter value
def num_squared(num, power=2):
    print(num ** power)
```

```python
num_squared(5,3) # This will print 125, which is 5 raised to the power of 3.
num_squared(5) # This will print 25, which is 5 squared, because the default value for power is 2.
```

### Arbitary Number of  Parameters

```python
# arbitrary number of parameters
def num_args(*numbers):
    for num in numbers:
        print(num)
```

```python
num_args(1, 2, 3, 4, 5) # This will print each number on a new line.
```

### Tuple for arbitary parameter

```python
num = (1, 2, 3, 4, 5)
def num_args1(*numbers):
    print(numbers[0] + numbers[1] )
```

```python
num_args1(num) #tuple index will be out of range because the function is expecting individual numbers, not a tuple. To fix this, you can unpack the tuple when calling the function:
```

```python
num_args1(*num) # This will unpack the tuple and pass each element as a separate argument to the function, resulting in the sum of the first two numbers, which is 3.
```

### Sequence of Parameters


```python
# change sequnce of parameters
def num_squared(num, power=2):
    print(num ** power)
```

```python
num_squared(4) # This will return 16, which is 4 squared.
num_squared(4,3) # This will return 64, which is 4 raised to the power of 3.
num_squared(power=3, num=4) # This will also return 64, because the parameters are specified by name, so the order does not matter.
```

### Keyword Arguments

```python
def number_kwarg(**numbers):
   print('my number is ' + numbers['integer'] + 'My other number is ' + numbers['integer2'])
```

```python
number_kwarg(integer='5', integer2='10') # This will print "my number is 5" and "My other number is 10", because the function is expecting keyword arguments with the keys 'integer' and 'integer2'.
```
