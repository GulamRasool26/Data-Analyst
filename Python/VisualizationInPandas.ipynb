# Pandas Visualizations

***Importing libraries***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

***Reading Data***

```python
df = pd.read_csv('Ice Cream Ratings.csv')
df
```

***Setting Index for the Dataframe***

```python
df = df.set_index('Date')
```

***plotting Default Kind is `Line plot`***

```python
df.plot()
```

***Changing the type of plot and adding subplots***

```python
df.plot(kind='bar', subplots=True)
```

```python
df.plot(kind='bar')
```

***Customizations of Labels***
1. Title
2. xlabels
3. ylabels

```python
df.plot(kind='bar', title='Ice Cream Ratings')
```

```python
df.plot(kind='bar', title='Ice Cream Ratings' , xlabel="Date",ylabel="Score")
```

```python
df.plot(kind='bar', title='Ice Cream Ratings' ,stacked=True, xlabel="Date",ylabel="Score")
```

```python
df.plot(kind='bar', title='Ice Cream Ratings' , xlabel="Date",ylabel="Score")
```

```python
df.plot.barh( title='Ice Cream Ratings' ,stacked=True, xlabel="Date",ylabel="Score")
```

```python
df.plot.scatter(x='Texture Rating', y='Overall Rating', s=100, c='Red')
```

```python
df.plot.hist(bins=20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize=(10,5))
```

```python
df.plot.pie(y='Flavor Rating', figsize=(5,5))
```
