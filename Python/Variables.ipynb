# Initializing multiple Variable Once

```python
x,y,z = "Chocolate", "Vanilla", "Strawberry"
print(x)
print(y)
print(z)
```

```python
x = y = z = "Chocolate"
print(x)
print(y)
print(z)
```

```python
ice_cream = ["Chocolate", "Vanilla", "Strawberry"]
x,y,z = ice_cream
print(x)
print(y)
print(z)
```

# Methods of typing Variables Name

```python
#Camel Case
myVariableName = "John"
print(myVariableName)
```

```python
#Pascal Case
MyVariableName = "John"
print(MyVariableName)
```

```python
#Snake Case
my_variable_name = "John"
print(my_variable_name)
```
