# **Web Scraping From a Real Website**

```python
# importing the necessary libraries
from bs4 import BeautifulSoup
import requests
```

***Requesting Webpages***

```python
# defining the URL of the webpage to scrape
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
```

```python
# user_agent is necessary to avoid being blocked by the website. It identifies the bot and provides contact information.
headers = {
    "User-Agent": "DataScienceLearningBot/1.0 (grmalik12353@gmail.com)"
}
response = requests.get(url, headers=headers)
```

```python
print(response.text)
```

***Parsing the response from Pages***

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup)
```

```python
<table class="wikitable sortable jquery-tablesorter">
```

***finding Table with class above from the HTML page***

```python
soup.find_all('table', class_='wikitable sortable')
```

```python
soup.find_all('table')[0]
```

```python
table = soup.find_all('table')[0]
table
```

***Got the Table Now Getting Column Names***

```python
world_titles = table.find_all('th')
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
```

```python
print(world_table_titles)
```

***Use Pandas to Create a DataFrame From HTML Tables***

```python
import pandas as pd
```

```python
pd.DataFrame(columns=world_table_titles)
```

```python
df = pd.DataFrame(columns=world_table_titles)
```

```python
df
```

***Separating Each row and Storing in a List***

```python
column_data = table.find_all('tr')
```

***Looping throught Each row and Separating cell data from Each row storing in DataFrame***

```python
for row in column_data[1:]:  # skipping the first row as it contains column names
    row_data = row.find_all("td")  # finding all the data in each row
    individual_row_data = [
        data.text.strip() for data in row_data
    ]  # extracting the text from each cell and stripping any extra whitespace
    df.loc[len(df)] = (
        individual_row_data  # adding the extracted data as a new row in the DataFrame
    )
```

```python
df
```

***Saving to CSV file***

```python
df.to_csv('largest_companies.csv', index=False)
```
