# Automatic File Sorter


```python
import os, shutil
```

```python
path = r"D:/Data-Analyst/Python/folder/"
```

```python
file_name = os.listdir(path)
```

```python
folder_name = [ 'xlsx files' ,'video files','pdf files', 'image files' , 'text files']
for loop in range(0,5):
    if not os.path.exists(path + folder_name[loop]):
        os.makedirs(path + folder_name[loop])
```

```python

for files in file_name:
    if ".xlsx" in files and not os.path.exists(path + "xlsx files/" + files):
        shutil.move(path + files , path + "xlsx files/"  + files)
    if ".png" in files and not os.path.exists(path + "image files/" + files):
        shutil.move(path + files , path + "image files/"+ files)
    if ".jpeg" in files and not os.path.exists(path + "image files/" + files):
        shutil.move(path + files , path + "image files/"+ files)
    if ".pdf" in files and not os.path.exists(path + "pdf files/" + files):
        shutil.move(path + files , path + "pdf files/"  + files)
    if ".MOV" in files and not os.path.exists(path + "video files/" + files):
        shutil.move(path + files , path + "video files/" + files)
```
